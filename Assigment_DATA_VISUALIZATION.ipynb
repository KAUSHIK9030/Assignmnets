```python
### 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.


import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a seed for reproducibility
np.random.seed(0)

# Generate a synthetic dataset with correlated features
size = 100  # Number of samples
feature_1 = np.random.normal(0, 1, size)
feature_2 = feature_1 * 0.8 + np.random.normal(0, 0.2, size)  # Correlated with feature_1
feature_3 = feature_1 * -0.5 + feature_2 * 0.5 + np.random.normal(0, 0.1, size)  # Correlated with both feature_1 and feature_2
feature_4 = np.random.normal(0, 1, size)  # Less correlated

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4
})

# Compute the correlation matrix
correlation_matrix = data.corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True)
plt.title('Correlation Matrix of Synthetic Dataset')
plt.show()
```

```python
### 4. Using the given x and y data, generate a 3D surface plot to visualize the function z=np.sin(np.sqrt(X**2 + Y**2))

# X = np.linspace(-5, 5, 100)
# Y = np.linspace(-5, 5, 100)
# X, Y = np.meshgrid(X,Y)

# data = {
#    'X' : x.flatten(),
#    'Y' : y.flatten(),
#    'Z' : z.flatten()
# }
# df = pd.DataFrame(data)



import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate X and Y arrays
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)

# Calculate Z values based on the function
Z = np.sin(np.sqrt(X**2 + Y**2))

# Flatten the arrays and create a DataFrame (optional for further analysis)
data = {
    'X': X.flatten(),
    'Y': Y.flatten(),
    'Z': Z.flatten()
}
df = pd.DataFrame(data)

# Plotting the 3D surface
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title('3D Surface Plot of $z = \sin(\sqrt{X^2 + Y^2})$')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Color bar
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
plt.show()



```

```python
### 1. create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.


from bokeh.plotting import figure, show, output_notebook, output_file
import numpy as np

# Display Bokeh plots inline
output_notebook()

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot with title and labels
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer
p.line(x, y, legend_label="sin(x)", line_width=2, color="navy")

# Save the plot as an HTML file (optional)
output_file("sine_wave.html")  # This will save the plot as "sine_wave.html"

# Show the plot
show(p)

```

```python
### 2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on th 'sizes' and 'colors' columns.


from bokeh.plotting import figure, show, output_notebook
from bokeh.io import output_file
import numpy as np
import pandas as pd

# Display plots inline
output_notebook()

# Generate random data
np.random.seed(42)
num_points = 100
data = {
    'x': np.random.rand(num_points) * 100,
    'y': np.random.rand(num_points) * 100,
    'sizes': np.random.randint(10, 30, num_points),
    'colors': np.random.choice(["#ff6347", "#6a5acd", "#3cb371", "#ffa07a", "#4682b4"], num_points)
}
df = pd.DataFrame(data)

# Create a new plot
p = figure(title="Random Scatter Plot with Sizes and Colors", x_axis_label='X', y_axis_label='Y')

# Add scatter plot
p.scatter('x', 'y', size='sizes', color='colors', source=df, alpha=0.6, legend_field='colors')

# Show the plot
show(p)
```

```python
### 3. Generate a bokeh bar chart representing the counts different fruits using the following dataset.

# fruits = ['Apples',  'Oranges', 'Bananas',  'Pears',]
# counts = [20, 25, 30, 35]


from bokeh.plotting import figure, show, output_notebook
from bokeh.transform import factor_cmap
from bokeh.io import output_file

# Display plots inline
output_notebook()

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a figure
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count', toolbar_location=None, tools="")

# Add bars
p.vbar(x=fruits, top=counts, width=0.5, color=factor_cmap('x', palette=['#ff6347', '#ffa07a', '#3cb371', '#6a5acd'], factors=fruits))

# Show plot
show(p)
```

```python
### 5. Create a Bokeh heatmap using the provided dataset.

# dataset_heatmap = np.random.rand(10,10)
# x = np.linspace(0, 1, 10)
# y = np.linspace(0, 1, 10)
# xx, yy = np.meshgrid(x, y)



from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import linear_cmap
import numpy as np

# Display plots inline
output_notebook()

# Generate data
dataset_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create the figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1), toolbar_location=None, tools="")

# Set up color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=dataset_heatmap.min(), high=dataset_heatmap.max())

# Plot the heatmap
p.image(image=[dataset_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)


```
