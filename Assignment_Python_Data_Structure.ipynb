##2. Explain the key features of lists in python.


lists in python are one of the most commonly used data structures. They are ordered, mutable and can contain elements of different data types.

**example:**
fruits = ["apple", "banana", "cherry"]

fruits[1]  # Output: banana

**(b) Mutable**  -  lists are mutable, meaning you can modify (add or remove or change) elements after the list is created.

**example:**

fruits = ["apple", "banana", "cherry"]


fruits[1] = "blueberry"

fruits   # Output: ['apple', 'blueberry', 'cherry']


**(c) Heterogeneous**  -  A list can hold elements of different data types, such as integers, strings, floats, or even other lists.


**example**

mixed_list = [1, "apple", 3.14, [2, 4, 6]]

mixed_list  # Output: [1, 'apple', 3.14, [2, 4, 6]]


**(d) Dynamic Size** -  Lists can grow or shrink dynamically. You can append, remove, or insert elements as needed without pre-defining the size.

**example**

numbers = [1, 2, 3]

numbers.append(4)

numbers  # Output: [1, 2, 3, 4]


**(e) Indexing & slicing** -   Lists support both positive and negative indexing, as well as slicing to access sub-lists.

numbers = [10, 20, 30, 40, 50]

print(numbers[2])    # Output: 30 (positive index)

print(numbers[-1])   # Output: 50 (negative index)

print(numbers[1:4])  # Output: [20, 30, 40] (slice from index 1 to 3)

-append(): Adds an element at the end of the list.

-insert(): Inserts an element at a specified index.

-remove(): Removes the first occurrence of a value.

-pop(): Removes an element by index (or the last

-sort(): Sorts the list in-place.

-reverse(): Reverses the elements of the list.

-extend(): Appends elements from another list (or any iterable).

-count(): Counts occurrences of an element.

**example**

print(fruits[0])  # Output: apple

**example**

fruits = ["apple", "banana", "cherry", "date"]

fruits[1] = "blueberry"

print(fruits)  # Output: ['apple', 'blueberry', 'cherry', 'date']

fruits[2:4] = ["kiwi", "orange"]

print(fruits)  # Output: ['apple', 'blueberry', 'kiwi', 'orange']


**(c) Deleting elements in a list** -  You can delete elements from a list using the del statement, the remove() method, or the pop() method.

*   Using del:
The del keyword removes elements by index or slices of elements from the list.

**example**

fruits = ["apple", "banana", "cherry", "date"]

print(fruits)  # Output: ['apple', 'cherry', 'date']


del fruits[1:3]

The remove() method removes the first occurrence of a specified value from the list.

**example**

##4. Compare and contrast tuples and lists with examples.

Tuples and lists are two fundamental data structures in Python. Both allow you to store collections of items, but they have key differences in terms of mutability, performance, and use cases. Here's a detailed comparison.

**(a) Mutability** -

*  Lists are mutable, meaning their elements can be changed (modified, added, or removed) after the list is created.
*  Tuples are immutable, meaning once a tuple is created, its elements cannot be changed, added, or removed.

**example**

fruits_list = ["apple", "banana", "cherry"]

fruits_list[1] = "blueberry"  # Modify the second element

print(fruits_list)  # Output: ['apple', 'blueberry', 'cherry']

# Tuple example (immutable)
fruits_tuple = ("apple", "banana", "cherry")
# fruits_tuple[1] = "blueberry"  # This will raise a TypeError


**(b) Syntax** -  

*  Lists are created using square brackets [].
*  Tuples are created using parentheses ().

**example**

# List
my_list = [1, 2, 3]

# Tuple
my_tuple = (1, 2, 3)


**(c) Use cases**  -  

*  Lists are used when you need a collection of items that may change over time (e.g., adding, removing, or modifying elements).
*  Tuples are used when you want to ensure that the data remains constant and does not need to change (e.g., storing fixed data like coordinates or configuration constants).

**example**

# List: Use when the data is expected to change

shopping_list = ["milk", "bread"]

shopping_list.append("eggs")

print(shopping_list)  # Output: ['milk', 'bread', 'eggs']

# Tuple: Can be Used when the data should remain unchanged

coordinates = (10, 20)

# coordinates[0] = 30  # This will raise a TypeError, as tuples are immutable



**(d) Performance** -  

*  Tuples are generally faster than lists because they are immutable. This allows Python to optimize the memory usage for tuples, making them more efficient in terms of performance.
*  Lists, being mutable, have a bit more overhead due to the flexibility they provide, which can make them slightly slower than tuples in some cases.

**example**

# You can use a tuple as a dictionary key
locations = {(10, 20): "Park", (30, 40): "Mall"}
# You cannot use a list as a dictionary key because lists are mutable


**(e) Methods and Operations**-  

*  Lists have many more built-in methods (like append(), remove(), pop(), clear(), etc.) due to their mutable nature.
*  Tuples have only a couple of methods, like count() and index(), since they are immutable.

**example**

# List methods
fruits_list = ["apple", "banana"]

fruits_list.append("cherry")

print(fruits_list)  # Output: ['apple', 'banana', 'cherry']

# Tuple methods
fruits_tuple = ("apple", "banana", "apple")

print(fruits_tuple.count("apple"))  # Output: 2

print(fruits_tuple.index("banana"))  # Output: 1


**(f) Iteration** -

*  Both lists and tuples can be iterated over in a for loop.
*  However, since tuples are immutable, they are often used in situations where you need to protect the data from modification during iteration.

**example**

# Iterating over a list
fruits_list = ["apple", "banana", "cherry"]

for fruit in fruits_list:
    print(fruit)

# Iterating over a tuple
fruits_tuple = ("apple", "banana", "cherry")

for fruit in fruits_tuple:
    print(fruit)









##5. Describe the key features of sets and provide examples of their use.

A set is an unordered, mutable collection of unique elements in Python. Sets are useful for storing items where duplicates are not allowed, and where the order of elements does not matter. They provide highly optimized operations for membership testing and eliminating duplicates. Below are the key features of sets along with examples of their use.

**(a) Unordered** -  Sets do not maintain any specific order for their elements. The items are stored in an arbitrary order, and you cannot rely on the order to be consistent.

**example**

fruits = {"apple", "banana", "cherry"}

print(fruits)  # Output: {'banana', 'apple', 'cherry'} (order may vary)

**(b) unique elements**  -  A set automatically eliminates duplicate values. If you add the same item multiple times, it will only appear once in the set.

**example**

numbers = {1, 2, 2, 3, 4, 4}

print(numbers)  # Output: {1, 2, 3, 4} (duplicates are removed)

**(c) Mutable** -  Sets are mutable, meaning you can add, remove, or modify elements after the set is created. However, only immutable types (like numbers, strings, and tuples) can be elements of a set.

**example**

fruits = {"apple", "banana"}

fruits.add("cherry")  # Adding an element

print(fruits)  # Output: {'apple', 'banana', 'cherry'}


**(d) Unindexed** - Sets do not support indexing, slicing, or other sequence-like behavior. You cannot access elements by an index because the elements are stored in no particular order.

**example**

fruits = {"apple", "banana", "cherry"}

# print(fruits[0])  # This will raise a TypeError because sets are unordered and unindexed



**(e) Efficient membership testing** - Sets are highly optimized for membership testing (i.e., checking if an element is in the set). The time complexity for membership testing is O(1) on average.

**example**

fruits = {"apple", "banana", "cherry"}

print("banana" in fruits)  # Output: True


**(f) No duplicates allowed** - if you try to add a duplicate element to a set, it will be ignored since sets inherently prevent duplicates.


**example**

fruits = {"apple", "banana"}

fruits.add("apple")

print(fruits)  # Output: {'apple', 'banana'} (duplicate 'apple' is ignored)


**(g) Frozen sets** -  Python provides a special type of set called frozen set, which is immutable. Once created, you cannot modify a frozen set (no adding or removing elements).

**example**

frozen_set = frozenset([1, 2, 3])

frozen_set.add(4)  # This will raise an AttributeError since frozen sets are immutable

print(frozen_set)  # Output: frozenset({1, 2, 3})


**(h) Use cases of sets**  -  Removing Duplicates from a List: Sets are great for eliminating duplicate values from a list.

numbers = [1, 2, 2, 3, 4, 4, 5]

unique_numbers = set(numbers)

print(unique_numbers)  # Output: {1, 2, 3, 4, 5}

#Set Membership Testing: Sets are optimal for checking if an item is part of a collection.

vowels = {'a', 'e', 'i', 'o', 'u'}

print('a' in vowels)  # Output: True

print('z' in vowels)  # Output: False

##8. Discuss the importance of dictionary keys being immutable and provide examples.

In Python, dictionary keys must be immutable. This is because dictionaries use a hashing mechanism to quickly look up values associated with a key, and mutable objects (like lists or dictionaries) can change, causing issues with this hashing process.



1. Hashing Requirement  -  If a key is mutable (i.e., it can be changed after being added to the dictionary), its hash value can change, breaking the mapping between the key and its value. This would make it impossible to find the value associated with that key.
2. Consistency of Keys  -  The immutability of keys ensures that once a key is added to a dictionary, it remains consistent and unchanging. If a key could be modified, it would disrupt the integrity of the dictionary, as the location where the value is stored would no longer be predictable.
3. Performance Considerations  -  The hashing mechanism relies on keys remaining constant so that the dictionary can perform fast lookups (with an average time complexity of O(1)). If keys were mutable, every time a key changed, the dictionary would have to rehash the key or rebuild its internal structure, leading to performance degradation.
